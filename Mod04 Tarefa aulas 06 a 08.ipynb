````python
#1) Seu código aqui

def recebe_estado(uf, estado):
    # Adiciona a sigla do Estado da url
    if estado != 'Estado':
        estado = 'Munic%C3%ADpio'
    url = f"https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={estado}"
    
    return url
uf = input('Qual o estado? ').upper().strip()

estado = str(input('Estado ou Município? ')).capitalize().strip()

r = recebe_estado(uf, estado)

url = (f'{r}')

r = requests.get(url)

data_json = r.json()

df = pd.DataFrame(data_json['items'])

df
````

````python
# 2) Seu código aqui]

uf = 'MG'

estado = 'Estado'

url = (f'{recebe_estado(uf, estado)}')

r = requests.get(url)

data_json = r.json()

df = pd.DataFrame(data_json['items'])

df[df['status'] == 'Arquivado por decurso de prazo'].value_counts('interessado')
````

````python
# 3) Seu código aqui
uf = 'BA'

estado = 'Municipio'

url = (f'{recebe_estado(uf, estado)}')

r = requests.get(url)

data_json = r.json()

df = pd.DataFrame(data_json['items'])

df[df['status'] == 'Deferido'].value_counts('interessado')
````

````python
# 4) Seu código aqui
uf = 'BA'

estado = 'Estado'

url = (f'{recebe_estado(uf, estado)}')

r = requests.get(url)

data_json = r.json()

df = pd.DataFrame(data_json['items'])

df_ba = pd.DataFrame(df['interessado'])

df_ba.to_csv('bahia.csv')
````
